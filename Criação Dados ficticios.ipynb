```python
# 1. Instalar as bibliotecas (rode esta célula se ainda não tiver instalado)
# !pip install Faker openpyxl

# 2. Importar as bibliotecas
import pandas as pd
from faker import Faker
import random
from datetime import datetime, timedelta

# Inicializar o gerador de dados fictícios em português do Brasil
fake = Faker('pt_BR')

# --- PARTE 1: GERAR A BASE DE CLIENTES ÚNICOS (df_clientes) ---
num_clientes_unicos = 500 # Defina quantos clientes únicos você quer ter
clientes_unicos = []

for _ in range(num_clientes_unicos):
    clientes_unicos.append({
        'ID_Cliente': random.randint(10000, 99999), # ID único para o cliente
        'Razao_Social': fake.company(),
        'Representante': fake.name(),
        'Cidade': fake.city(),
        'Estado': fake.state_abbr(),
        'CNPJ': fake.cnpj(),
        'Tipo_Cliente': random.choice(['PF', 'PJ']),# Exemplo: Pessoa Física ou Jurídica
        'E-mail': fake.email(),
        'Telefone': fake.phone_number(),
        'Celular': fake.phone_number(),
    })

df_clientes = pd.DataFrame(clientes_unicos)

print("---")
print("Primeiras 5 linhas do DataFrame de Clientes Únicos (df_clientes):")
print(df_clientes.head())
print("\nInformações do DataFrame de Clientes:")
print(df_clientes.info())

# --- PARTE 2: GERAR UMA TABVELA DE PRODUTOS(df_produtos)--

num_produtos = 6

dados_produtos = []

for i in range(num_produtos):
    cod_produto = ''.join(random.choices('ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k=random.randint(6, 12))),
    desc_produto = random.choice(['EVAPORADOR', 'TROCADOR DE CALOR', 'KIT DEGELO', 'KIT VENTILACAO', 'COMPRESSOR', 'CONDENSADOR'])


dados_produtos.append({
    'Cod_Produto': cod_produto,
    'Desc_Produto': desc_produto
})

df_produtos = pd.DataFrame(dados_produtos)



# --- PARTE 3: GERAR A BASE DE PEDIDOS (df_pedidos) ---
num_pedidos_total = 10000 # Defina quantos pedidos você quer gerar no total

dados_pedidos = []

for i in range(num_pedidos_total):
    # Selecionar um cliente aleatório da base de clientes únicos
    # Isso garante que cada pedido esteja associado a um cliente existente
    cliente_do_pedido = df_clientes.sample(1).iloc[0]

    #Selecionar os produtos com seus IDS e seus codigos e nomes
    produto_do_pedido = df_produtos.sample(1).iloc[0]


    # Atribuir o ID do cliente e outras informações relevantes do cliente ao pedido
    id_cliente_pedido = cliente_do_pedido['ID_Cliente']
    razao_social_pedido = cliente_do_pedido['Razao_Social'] # Repetindo para facilidade de visualização
    cnpj_pedido = cliente_do_pedido['CNPJ'] # Repetindo para facilidade de visualização

    # Gerar dados específicos do pedido
    cod_item = random.randint(10000, 999999) # Código do item varia por pedido
    cod_produto = produto_do_pedido['Cod_Produto']
    desc_produto = produto_do_pedido['Desc_Produto']
    quantidade = random.randint(1, 10)
    preco_unitario = round(random.uniform(10.00, 1000.00), 2)
    valor_total_item = round(quantidade * preco_unitario, 2)

    # Gerar datas do pedido
    data_emissao_pedido = fake.date_between(start_date='-2y', end_date='today')
    data_desejada_pedido = data_emissao_pedido + timedelta(days=random.randint(5, 60))
    data_firmada_original_pedido = data_desejada_pedido + timedelta(days=random.randint(0, 15))
    data_firmada_atual_pedido = data_firmada_original_pedido
    data_final_pedido = data_firmada_atual_pedido + timedelta(days=random.randint(0, 30))

    # Dados da Nota Fiscal e Embarque (associados a este pedido específico)
    nf = random.randint(100000, 999999)
    data_nf = data_emissao_pedido + timedelta(days=random.randint(1, 10))
    data_nf_oficial = data_nf
    data_embarque = data_nf + timedelta(days=random.randint(1, 5))
    data_emb_oficial = data_embarque

    tipo_fat = random.choice(['VENDA', 'DEVOLUCAO', 'BONIFICACAO', 'AMOSTRA'])
    transportadora = fake.company() + " TRANSPORTES LTDA" # Pode ser a mesma do cliente ou diferente
    cond_pagamento = random.choice(['BOLETO DESCONTADO', 'CARTEIRA A VISTA', 'CARTEIRA A PRAZO', 'CARTAO DE CREDITO', 'PIX', 'TRANSFERENCIA'])
    valor_geral_pedido = valor_total_item # Valor total do pedido

    dados_pedidos.append({
        'ID_Pedido': i + 1, # ID único para cada pedido
        'ID_Cliente': id_cliente_pedido, # Chave para conectar com df_clientes
        'Razao_Social_Cliente': razao_social_pedido, # Para facilitar a visualização no df_pedidos
        'CNPJ_Cliente': cnpj_pedido, # Para facilitar a visualização no df_pedidos
        'Representante': cliente_do_pedido['Representante'], # Repetindo para facilidade de visualização
        'Cod_Produto': cod_produto,
        'Desc_Produto': desc_produto,
        'Quantidade': quantidade,
        'Preco_Unitario': preco_unitario,
        'Valor_Total_Item': valor_total_item,
        'NF': nf,
        'Data_Emissao_Pedido': data_emissao_pedido,
        'Data_Desejada_Entrega': data_desejada_pedido,
        'Data_Firmada_Original': data_firmada_original_pedido,
        'Data_Firmada_Atual': data_firmada_atual_pedido,
        'Data_Final_Entrega': data_final_pedido,
        'Data_NF': data_nf,
        'Data_NF_Oficial': data_nf_oficial,
        'Data_Embarque': data_embarque,
        'Data_Emb_Oficial': data_emb_oficial,
        'Tipo_Faturamento': tipo_fat,
        'Transportadora': transportadora,
        'Condicao_Pagamento': cond_pagamento,
        'Valor_Geral_Pedido': valor_geral_pedido
    })

df_pedidos = pd.DataFrame(dados_pedidos)

print("---")
print("Primeiras 5 linhas do DataFrame de Pedidos (df_pedidos):")
print(df_pedidos.head())
print("\nInformações do DataFrame de Pedidos:")
print(df_pedidos.info())



# Opcional: Salvar como Excel
# with pd.ExcelWriter('dados_ficticios_relacionados.xlsx') as writer:
#     df_clientes.to_excel(writer, sheet_name='Clientes', index=False)
#     df_pedidos.to_excel(writer, sheet_name='Pedidos', index=False)
# print(f"As duas bases salvas em 'dados_ficticios_relacionados.xlsx' como abas separadas!")
```
